# Plot distribution of the match's duration vs walk distance one below the other.

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
p1 = pubg['matchDuration']
p2 = pubg['walkDistance']
plt.subplot(2, 1, 1)
plt.plot(p1)
plt.title('matchDuration')
plt.subplot(2, 1, 2)
plt.plot(p2)
plt.xlabel('walkDistance')
plt.show()




```

# Plot distribution of the match's duration vs walk distance side by side.

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
p1 = pubg['matchDuration']
p2 = pubg['walkDistance']
plt.subplot(1, 2, 1)
plt.plot(p1)
plt.title('matchDuration')

plt.subplot(1, 2, 2)
plt.plot(p2)
plt.title('walkDistance')
plt.show()
```
